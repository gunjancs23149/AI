```python
isa = {
    "bird": "animal",
    "dog": "animal",
    "sparrow": "bird"
}
has = {
    "animal": "cells",
    "dog": "tail"
}
can = {
    "bird": "fly",
    "dog": "bark"
}
def check_isa(x):
    if x in isa:
        print(x, "is a", isa[x])
    else:
        print("No ISA relation found")
def check_has(x):
    if x in has:
        print(x, "has", has[x])
    else:
        print("No HAS relation found")
def check_can(x):
    if x in can:
        print(x, "can", can[x])
    else:
        print("No CAN relation found")
check_isa("sparrow")
check_has("dog")
check_can("bird")
```

```python
#pr9
P_cloudy = 0.40
P_rain = 0.20
P_cloudy_given_rain = 0.85
P_rain_given_cloudy = (P_rain * P_cloudy_given_rain) / P_cloudy
print("Probability of Rain given Cloudy Weather:")
print("P(Rain | Cloudy) =", P_rain_given_cloudy)

print("\nPercentage Probability:")
print("P(Rain | Cloudy) =", P_rain_given_cloudy * 100, "%")
```

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

path = "/content/customer_shopping_data.csv"
df = pd.read_csv(path)

encoder = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = encoder.fit_transform(df[col])

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(accuracy)

cm = confusion_matrix(y_test, y_pred)
print(cm)

precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(precision)
print(recall)
print(f1)
```

```python
%%time

# The existing code from cell 7FGC9y-p7FmT that is likely taking a long time.
# This includes data loading, preprocessing, train-test split, model training, and prediction.

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score

path = "/content/customer_shopping_data.csv"
df = pd.read_csv(path)

# Sampling a smaller portion of the data to reduce execution time
print(f"Original dataset size: {len(df)} rows")
df = df.sample(frac=0.1, random_state=42) # Using 10% of the data
print(f"Reduced dataset size after sampling: {len(df)} rows")

encoder = LabelEncoder()
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = encoder.fit_transform(df[col])

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = SVC()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{cm}")

precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```
